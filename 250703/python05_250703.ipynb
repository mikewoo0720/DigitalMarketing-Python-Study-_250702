- 구구단을 2단부터 9단까지 다음과 같이 출력해주세요!!!
  ```
  2 * 1 = 2
  2 * 2 = 4
  ...
  2 * 9 = 18
  3 * 1 = 3
  3 * 2 = 6
  ...
  9 * 9 = 81

  -for문을 선언하여 두가지 정수 타입을 불러오고
  -불러온 정수 타입을 곱하기 한 값을 출력되는 형태

````python
for mul1 in range(1, 10) :
    for mul2 in range(1, 10):
        print(mul1*mul2)
````

````python
  # -for문을 두번 선언하여 두가지 정수 타입을 불러오고
  # -불러온 정수 타입을 곱하기 한 값을 출력되는 형태

for digit in range(2, 10) :
    for digit1 in range(1, 10) :
        print(f"{digit} * {digit1} = {digit * digit1}")
````
